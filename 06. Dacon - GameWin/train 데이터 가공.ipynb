```python
import pandas as pd
```

```python
train = pd.read_csv('./train.csv')
train
```

```python
target = train['winner']
data = train.drop('winner', axis=1)
```

```python
print(target.shape)
print(data.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.01, shuffle=True, stratify=target)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
print('dataframe')
print('원본 : ', data.shape, '가공 : ', X_test.shape)

print()

for col in data.columns:
    print(col)
    print('원본 : ', len(data[col].unique()), '가공 : ', len(X_test[col].unique()))
```

```python
result = pd.concat([y_test, X_test], axis=1)
result.to_csv('./가공train.csv', index=False)
```
